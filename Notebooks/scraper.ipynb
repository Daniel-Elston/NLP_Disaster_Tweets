```python
import requests
from bs4 import BeautifulSoup

def scrape_news_website(url):
    # Send a GET request to the website
    response = requests.get(url)
    response.raise_for_status()

    # Create a BeautifulSoup object and parse the HTML content
    soup = BeautifulSoup(response.content, 'html.parser')

    # Find the relevant HTML elements containing the headlines and URLs
    headlines = soup.find_all('h3')
    urls = soup.find_all('a', href=True)

    # Extract the text and URLs
    news_data = []
    for headline, url in zip(headlines, urls):
        news_data.append({
            'headline': headline.text.strip(),
            'url': url['href']
        })

    return news_data

# Example usage
news_website_url = 'https://www.bbc.co.uk/news/business'
scraped_data = scrape_news_website(news_website_url)

# Print the scraped data
for data in scraped_data:
    print(data['headline'])
    print(data['url'])
    print()
```
